````python
# Data preparation
nav_data['date'] = pd.to_datetime(nav_data['date'])
nav_data = nav_data.sort_values('date')

# Create plot with all 40 schemes
fig, ax = plt.subplots(figsize=(16, 8))

# Plot each scheme with semi-transparent lines
for scheme in nav_data['scheme_name'].unique():
    scheme_data = nav_data[nav_data['scheme_name'] == scheme]
    ax.plot(scheme_data['date'], scheme_data['nav'], alpha=0.4, linewidth=0.8)

# Highlight 2023 bull run (green background)
ax.axvspan(pd.Timestamp('2023-01-01'), pd.Timestamp('2023-12-31'), 
           alpha=0.15, color='green', label='2023 Bull Run')

# Highlight 2024 corrections (red background)
ax.axvspan(pd.Timestamp('2024-01-01'), pd.Timestamp('2024-12-31'), 
           alpha=0.15, color='red', label='2024 Market Corrections')

# Formatting
ax.set_xlabel('Date', fontsize=12, fontweight='bold')
ax.set_ylabel('NAV (₹)', fontsize=12, fontweight='bold')
ax.set_title('NAV Trends Across 40 Mutual Fund Schemes (2022–2026)\nHighlighting 2023 Bull Run & 2024 Corrections', 
             fontsize=14, fontweight='bold', pad=20)
ax.legend(loc='upper left', fontsize=11)
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig(f'{charts_dir}/01_nav_trends_2022_2026.png', dpi=300, bbox_inches='tight')
plt.show()

print("✓ Chart 1: NAV trends exported")
````

## 2. AUM Growth by Fund House (2022–2025)

**Analysis**: Grouped bar chart comparing AUM for each fund house across years. Highlights SBI's ₹12.5L Cr dominance.

**Key Insight**: SBI consistently leads the mutual fund industry in asset under management.

````python
# Data preparation
aum_pivot = aum_data.pivot_table(values='aum_crores', index='fund_house', columns='year', aggfunc='sum')

# Create grouped bar chart
fig, ax = plt.subplots(figsize=(14, 8))

x = np.arange(len(aum_pivot.index))
width = 0.2
years = [2022, 2023, 2024, 2025]
colors = sns.color_palette('Set2', len(years))

for i, year in enumerate(years):
    if year in aum_pivot.columns:
        values = aum_pivot[year].values
        ax.bar(x + i * width, values, width, label=str(year), color=colors[i])

# Highlight SBI at ₹12.5L Cr (if present)
if 'SBI' in aum_pivot.index:
    sbi_idx = list(aum_pivot.index).index('SBI')
    sbi_2025 = aum_pivot.loc['SBI', 2025] if 2025 in aum_pivot.columns else 0
    ax.text(sbi_idx + width * 1.5, sbi_2025 + 20000, 
           f'₹{sbi_2025/100000:.1f}L Cr', ha='center', fontweight='bold', color='darkred', fontsize=10)

# Formatting
ax.set_xlabel('Fund House', fontsize=12, fontweight='bold')
ax.set_ylabel('AUM (₹ Crores)', fontsize=12, fontweight='bold')
ax.set_title('AUM Growth by Fund House (2022–2025)\nSBI\'s Dominance in Indian Mutual Funds', 
             fontsize=14, fontweight='bold', pad=20)
ax.set_xticks(x + width * 1.5)
ax.set_xticklabels(aum_pivot.index, rotation=45, ha='right')
ax.legend(title='Year', fontsize=10)
ax.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.savefig(f'{charts_dir}/02_aum_by_fund_house.png', dpi=300, bbox_inches='tight')
plt.show()

print("✓ Chart 2: AUM growth by fund house exported")
````

## 3. SIP Inflow Time-Series (Jan 2022 – Dec 2025)

**Analysis**: Monthly SIP trend line chart annotating the ₹31,002 Cr all-time high in December 2025.

**Key Insight**: SIP inflows show steady growth with seasonal patterns and peak in Dec 2025.

````python
# Data preparation
sip_data['date'] = pd.to_datetime(sip_data['date'])
sip_data = sip_data.sort_values('date')
sip_monthly = sip_data.groupby(sip_data['date'].dt.to_period('M'))['sip_inflow'].sum()
sip_monthly.index = sip_monthly.index.to_timestamp()

# Create time-series plot
fig, ax = plt.subplots(figsize=(16, 8))

ax.plot(sip_monthly.index, sip_monthly.values, linewidth=2.5, marker='o', markersize=5, color='#2E86AB')
ax.fill_between(sip_monthly.index, sip_monthly.values, alpha=0.3, color='#2E86AB')

# Annotate all-time high (Dec 2025)
max_idx = sip_monthly.idxmax()
max_val = sip_monthly.max()
ax.annotate(f'All-time High\n₹{max_val:,.0f} Cr\n(Dec 2025)', 
            xy=(max_idx, max_val), xytext=(max_idx - pd.Timedelta(days=120), max_val + 2000),
            fontsize=11, fontweight='bold', color='darkred',
            bbox=dict(boxstyle='round,pad=0.8', facecolor='yellow', alpha=0.7),
            arrowprops=dict(arrowstyle='->', color='darkred', lw=2))

# Formatting
ax.set_xlabel('Month', fontsize=12, fontweight='bold')
ax.set_ylabel('SIP Inflow (₹ Crores)', fontsize=12, fontweight='bold')
ax.set_title('Monthly SIP Inflows (Jan 2022 – Dec 2025)\nSteady Growth with Peak Performance in December 2025', 
             fontsize=14, fontweight='bold', pad=20)
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig(f'{charts_dir}/03_sip_inflows_timeseries.png', dpi=300, bbox_inches='tight')
plt.show()

print(f"✓ Chart 3: SIP inflows time-series exported")
print(f"  Peak SIP inflow: ₹{max_val:,.0f} Cr on {max_idx.strftime('%B %Y')}")
````

## 4. Category Inflow Heatmap

**Analysis**: Seaborn heatmap showing net inflows by fund category and month (2022–2025).

**Key Insight**: Identifies seasonal patterns and category-specific inflow trends.

````python
# Data preparation
investor_data['date'] = pd.to_datetime(investor_data['date'])
investor_data['year_month'] = investor_data['date'].dt.to_period('M')

# Create heatmap data (net inflow by category and month)
heatmap_data = investor_data.pivot_table(
    values='sip_amount', index='fund_category', columns='year_month', aggfunc='sum'
)

# Create heatmap
fig, ax = plt.subplots(figsize=(18, 8))

sns.heatmap(heatmap_data, cmap='YlGnBu', annot=False, fmt='.0f', 
            cbar_kws={'label': 'Net Inflow (₹ Crores)'}, ax=ax, linewidths=0.5)

# Formatting
ax.set_xlabel('Month', fontsize=12, fontweight='bold')
ax.set_ylabel('Fund Category', fontsize=12, fontweight='bold')
ax.set_title('Category-wise Inflow Trends Heatmap (2022–2025)\nMonths × Fund Categories', 
             fontsize=14, fontweight='bold', pad=20)
ax.set_xticklabels([label.get_text() for label in ax.get_xticklabels()], rotation=45, ha='right')

plt.tight_layout()
plt.savefig(f'{charts_dir}/04_category_inflow_heatmap.png', dpi=300, bbox_inches='tight')
plt.show()

print("✓ Chart 4: Category inflow heatmap exported")
````

## 5. Investor Demographics

**Analysis**: 
- Age group distribution pie chart
- SIP amount box plot by age group
- Gender split pie chart

**Key Insight**: Identifies demographic trends in SIP investor base.

````python
# Create 3-panel demographic visualization
fig = plt.figure(figsize=(16, 5))

# Panel 1: Age group distribution (pie chart)
ax1 = plt.subplot(1, 3, 1)
age_dist = investor_data['age_group'].value_counts()
colors_age = sns.color_palette('pastel', len(age_dist))
ax1.pie(age_dist.values, labels=age_dist.index, autopct='%1.1f%%', colors=colors_age, startangle=90)
ax1.set_title('Age Group Distribution\n(Investor Base)', fontsize=12, fontweight='bold')

# Panel 2: SIP amount box plot by age group
ax2 = plt.subplot(1, 3, 2)
investor_data.boxplot(column='sip_amount', by='age_group', ax=ax2)
ax2.set_xlabel('Age Group', fontsize=11, fontweight='bold')
ax2.set_ylabel('SIP Amount (₹)', fontsize=11, fontweight='bold')
ax2.set_title('SIP Amount Distribution\nby Age Group', fontsize=12, fontweight='bold')
plt.sca(ax2)
plt.xticks(rotation=45)

# Panel 3: Gender split (pie chart)
ax3 = plt.subplot(1, 3, 3)
gender_dist = investor_data['gender'].value_counts()
colors_gender = ['#FF69B4', '#4169E1']
ax3.pie(gender_dist.values, labels=gender_dist.index, autopct='%1.1f%%', colors=colors_gender, startangle=90)
ax3.set_title('Gender Distribution\n(Investor Base)', fontsize=12, fontweight='bold')

plt.suptitle('Investor Demographics Analysis', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig(f'{charts_dir}/05_investor_demographics.png', dpi=300, bbox_inches='tight')
plt.show()

print("✓ Chart 5: Investor demographics exported")
print(f"\nAge group breakdown:\n{age_dist}")
print(f"\nGender breakdown:\n{gender_dist}")
````

## 7. Folio Count Growth (2022–2025)

**Analysis**: Line chart tracking folio count from 13.26 Cr (Jan 2022) to 26.12 Cr (Dec 2025). Key milestones marked.

**Key Insight**: Shows near doubling of investor folios over 4 years.

````python
# Data preparation
folio_data['date'] = pd.to_datetime(folio_data['date'])
folio_data = folio_data.sort_values('date')

# Create line chart
fig, ax = plt.subplots(figsize=(16, 8))

ax.plot(folio_data['date'], folio_data['folio_count_crores'], linewidth=3, marker='o', markersize=6, color='#E63946')
ax.fill_between(folio_data['date'], folio_data['folio_count_crores'], alpha=0.2, color='#E63946')

# Mark key milestones
milestones = [
    (folio_data.iloc[0], '13.26 Cr\n(Jan 2022)'),
    (folio_data.iloc[len(folio_data)//2], 'Mid-point\n(2023)'),
    (folio_data.iloc[-1], '26.12 Cr\n(Dec 2025)')
]

for data_point, label in milestones:
    ax.annotate(label, xy=(data_point['date'], data_point['folio_count_crores']),
               xytext=(0, 10), textcoords='offset points', ha='center',
               fontsize=10, fontweight='bold',
               bbox=dict(boxstyle='round,pad=0.5', facecolor='lightyellow', alpha=0.8))

# Formatting
ax.set_xlabel('Date', fontsize=12, fontweight='bold')
ax.set_ylabel('Folio Count (Crores)', fontsize=12, fontweight='bold')
ax.set_title('Folio Count Growth (2022–2025)\nNear Doubling of Investor Base over 4 Years', 
             fontsize=14, fontweight='bold', pad=20)
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig(f'{charts_dir}/07_folio_count_growth.png', dpi=300, bbox_inches='tight')
plt.show()

print("✓ Chart 7: Folio count growth exported")
growth_pct = ((folio_data.iloc[-1]['folio_count_crores'] - folio_data.iloc[0]['folio_count_crores']) / folio_data.iloc[0]['folio_count_crores']) * 100
print(f"  Folio count growth: {growth_pct:.1f}% over 4 years")
````

## 8. NAV Return Correlation Matrix

**Analysis**: Compute pairwise correlation of daily returns for 10 selected schemes. Seaborn heatmap visualization.

**Key Insight**: Identifies relationships between fund performance and correlation patterns.

````python
# Select top 10 funds by AUM
top_10_funds = nav_data.groupby('scheme_name')['nav'].count().nlargest(10).index.tolist()

# Calculate daily returns
returns_data = {}
for fund in top_10_funds:
    fund_nav = nav_data[nav_data['scheme_name'] == fund].sort_values('date')
    returns_data[fund] = fund_nav['nav'].pct_change().values

# Create returns DataFrame
returns_df = pd.DataFrame(returns_data)

# Compute correlation matrix
corr_matrix = returns_df.corr()

# Create heatmap
fig, ax = plt.subplots(figsize=(12, 10))

sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0,
            square=True, linewidths=1, cbar_kws={'label': 'Correlation Coefficient'}, ax=ax)

ax.set_title('NAV Daily Returns Correlation Matrix\n(Top 10 Schemes)', 
             fontsize=14, fontweight='bold', pad=20)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

plt.tight_layout()
plt.savefig(f'{charts_dir}/08_nav_correlation_matrix.png', dpi=300, bbox_inches='tight')
plt.show()

print("✓ Chart 8: NAV correlation matrix exported")
print(f"\nAverage correlation: {corr_matrix.values[np.triu_indices_from(corr_matrix.values, k=1)].mean():.3f}")
````

## 9. Sector Allocation Donut Chart

**Analysis**: Aggregate sector weights from portfolio holdings across all equity funds.

**Key Insight**: Shows portfolio diversification and sector concentration.

````python
# Aggregate sector allocation
sector_allocation = holdings_data.groupby('sector')['weight_percent'].sum().sort_values(ascending=False)

# Create donut chart
fig, ax = plt.subplots(figsize=(12, 8))

colors = sns.color_palette('Set3', len(sector_allocation))
wedges, texts, autotexts = ax.pie(sector_allocation.values, labels=sector_allocation.index, 
                                    autopct='%1.1f%%', colors=colors, startangle=90)

# Create donut effect
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
ax.add_artist(centre_circle)

# Formatting
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontweight('bold')
    autotext.set_fontsize(9)

ax.set_title('Sector Allocation Across All Equity Funds\n(Donut Chart)', 
             fontsize=14, fontweight='bold', pad=20)

plt.tight_layout()
plt.savefig(f'{charts_dir}/09_sector_allocation_donut.png', dpi=300, bbox_inches='tight')
plt.show()

print("✓ Chart 9: Sector allocation donut exported")
print(f"\nTop 5 sectors by weight:\n{sector_allocation.head(5)}")
````

---
